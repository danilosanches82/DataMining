### Histograma das classes

### Histograma dos atributos - Base Wine

```python
from sklearn.datasets import load_wine
from matplotlib import pyplot
import pandas
#dados, target = load_diabetes(return_X_y=True)
Data = load_wine()

f, ax = pyplot.subplots(3,4)
ax = ax.flatten()
tam = len(Data.feature_names)

dados = pandas.DataFrame(Data.data, columns=Data.feature_names)

for column, posicao in zip(dados, range(0, tam-1, 1)) :
    ax[posicao].hist(dados[column].dropna(), bins='auto')
    ax[posicao].set_title('Histograma')
    ax[posicao].set_ylabel('Frequencia')
    ax[posicao].set_xlabel(column)

pyplot.tight_layout()    
pyplot.show()    
```

### Scatter plot

```python
from sklearn.datasets import load_wine
from pandas.plotting import scatter_matrix
from matplotlib import pyplot
import pandas

Data = load_wine()
target = Data.target

dados = pandas.DataFrame(Data.data, columns=Data.feature_names)

# criando um dicionario para mapear cada classe para uma cor
classe_cor = {0 : 'red', 
              1 : 'blue', 
              2 : 'yellow'
              }

# criando uma lista com as cores de cada exemplo
cores = [classe_cor[nome] for nome in target]# gerando scatter plot

#print(cores)

pyplot.scatter(dados['alcohol'], dados['malic_acid'], alpha=0.2,
            c=cores)
pyplot.xlabel(Data.feature_names[0])
pyplot.ylabel(Data.feature_names[1]);
pyplot.show()
```

#### Rótulos sem formatação 

```python
# calculando o scatter_Matrix para base Wine
from sklearn.datasets import load_wine
from pandas.plotting import scatter_matrix
from matplotlib import pyplot
import pandas

Data = load_wine()
target = Data.target

tam = len(Data.feature_names)

dados = pandas.DataFrame(Data.data, columns=Data.feature_names)

# criando um dicionario para mapear cada classe para uma cor
classe_cor = {0 : 'red', 
              1 : 'blue', 
              2 : 'yellow'
              }

# criando uma lista com as cores de cada exemplo
cores = [classe_cor[nome] for nome in target]

scatter_matrix(dados, alpha=0.2, figsize=(20, 20), diagonal='kde', color = cores)

pyplot.show()
```

#### Rótulos com rotação

```python
# calculando o scatter_Matrix para base Diabetes
from sklearn.datasets import load_wine
from pandas.plotting import scatter_matrix
from matplotlib import pyplot
import pandas

Data = load_wine()
target = Data.target

tam = len(Data.feature_names)

dados = pandas.DataFrame(Data.data, columns=Data.feature_names)

# criando um dicionario para mapear cada classe para uma cor
classe_cor = {0 : 'red', 
              1 : 'blue', 
              2 : 'yellow'
              }

# criando uma lista com as cores de cada exemplo
cores = [classe_cor[nome] for nome in target]

sm = scatter_matrix(dados, alpha=0.2, figsize=(20, 20), diagonal='kde', color = cores)

#Change label rotation
[s.xaxis.label.set_rotation(45) for s in sm.reshape(-1)]
[s.yaxis.label.set_rotation(0) for s in sm.reshape(-1)]

#May need to offset label when rotating to prevent overlap of figure
[s.get_yaxis().set_label_coords(-0.3,0.5) for s in sm.reshape(-1)]

#Hide all ticks
[s.set_xticks(()) for s in sm.reshape(-1)]
[s.set_yticks(()) for s in sm.reshape(-1)]
pyplot.show()
```

#### Renomeando os rótulos

```python
# calculando o scatter_Matrix para base Diabetes
from sklearn.datasets import load_wine
from pandas.plotting import scatter_matrix
from matplotlib import pyplot
import pandas

Data = load_wine()
target = Data.target

tam = len(Data.feature_names)

dados = pandas.DataFrame(Data.data, columns=Data.feature_names)

# renomeando os atributos
dados.columns = ['A', 'B', 'C', 'D', 'E',
       'F', 'G', 'H', 'I', 'J', 'K',
       'L', 'M']

# criando um dicionario para mapear cada classe para uma cor
classe_cor = {0 : 'red', 
              1 : 'blue', 
              2 : 'yellow'
              }

# criando uma lista com as cores de cada exemplo
cores = [classe_cor[nome] for nome in target]

scatter_matrix(dados, alpha=0.2, figsize=(20, 20), diagonal='kde', color = cores)

pyplot.show()
```
