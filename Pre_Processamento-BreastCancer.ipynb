### Dados Faltantes

Missing Data ou Missing Values s√£o valores ausentes mostrados como NaN que significa "Not a Number" e s√£o do tipo float. Ter missing values no dataset √© um problema porque eles podem influenciar nas conclus√µes sobre os dados. Isto porque os missing values s√£o omitidos nas estat√≠sticas e s√£o ignorados nas opera√ß√µes de count, sum, etc. Al√©m disso, os valores ausentes reduzem o poder estat√≠stico, podem enviesar a estimativa dos par√¢metros, reduzir a representatividade das amostras, distorcer os resultados das an√°lises e gerar conclus√µes inv√°lidas.

A melhor forma de evitar os missing values √© selecionar os registros mais completos no dataset de origem. Por√©m, nem sempre isso √© poss√≠vel porque simplesmente as vezes a informa√ß√£o n√£o est√° dispon√≠vel. Nestes casos, as principais t√©cnicas para lidar com missing data s√£o:

Inser√ß√£o: preencher com os valores faltantes, inserir a m√©dia ou mediana ou moda dos valores (para num√©ricos), usar um modelo preditivo para estimar os valores faltantes ou simplesmente informar que √© um valor Desconhecido (para dados categ√≥ricos e textuais).
Exclus√£o Parcial: excluir os registros que est√£o com valores faltantes.

#### A sintaxe Python usada:

- Para selecionar NaN: pd.isnull()
- Para substituir valores NaN: df.fillna()
- Para substituir NaN pela m√©dia: df.fillna(df.mean(), inplace=True)
- Para substituir valores presentes por outros valores: `pd.replace("valor","novo_valor")
- Para remover os registros NaN: df.dropna(inplace=True) 

```python
import pandas
# carrega uma versao do conjunto de dados contendo valores
# ausentes
dados = pandas.read_csv('breast_cancer_missing.csv')

print(dados.isnull().sum())

valores_ausentes = dados.isnull()
medias = dados.mean()
dados.fillna(medias, inplace=True)
print(valores_ausentes)
print(medias)
print(dados.isnull().sum())
```

### Dados Faltantes com SimpleImputer

```python
import pandas
import numpy
from sklearn.impute import SimpleImputer

# carrega uma versao do conjunto de dados contendo valores
# ausentes
dados = pandas.read_csv('breast_cancer_missing.csv')

previsores = dados.iloc[:,0:31].values
classe = dados.iloc[:, 31].values

print(previsores)

imputer = SimpleImputer(missing_values=numpy.nan, strategy='mean')
imputer = imputer.fit(previsores)
previsores = imputer.transform(previsores)
```

```python
import pandas
# carregando uma versao do conjunto de dados contendo
# objetos redundantes (repetidos)
dados = pandas.read_csv('breast_cancer_red.csv')
# removendo exemplos redundantes
# observe que o resultado sera a base original
# o parametro inplace determina se a alteracao deve
# ocorrer no proprio DataFrame
# portanto, caso inplace=False, a funcao drop_duplicates
# retorna uma copia de breast_cancer_red com os exemplos
# redundantes removidos
dados.drop_duplicates(inplace=False)
print(dados)
```

### Normaliza√ß√£o de atributos num√©ricos

A normaliza√ß√£o por reescala define, atrav√©s de um valor m√≠nimo e um valor m√°ximo, um novo intervalo onde os valores de um atributo estar√£o contidos. Tipicamente, tal intervalo √© definido como [0, 1]. Portanto, para este caso, a normaliza√ß√£o por reescala de um atributo $j$ de um objeto $x_i$ pode ser calculada como:

$$ x_{ij} = \frac{x_{ij} - min_j}{max_j - min_j} $$

sendo min_j e max_j , nessa ordem, os valores m√≠nimo e m√°ximo do atributo $j$ para o conjunto de dados considerado.

Na normaliza√ß√£o por padroniza√ß√£o, os diferentes atributos cont√≠nuos poder√£o abranger diferentes intervalos, mas dever√£o possuir os mesmos valores para alguma medida de posi√ß√£o e de espalhamento/varia√ß√£o. 
Essas medidas ir√£o consistir na m√©dia e no desvio-padr√£o. Neste caso, o valor normalizado de um atributo $j$
em um objeto $i$ √© dado por:

$$ x_{ij} = \frac{x_{ij} - \bar{x}_j}{\sigma_j} $$

https://www.scalr.ai/pandas/normalize-column-pandas-dataframe

```python
from sklearn.preprocessing import MaxAbsScaler

# split in predictors and class
X = dados.iloc[:, 2:31].values
Y = dados.iloc[:, 1].values

print('original data:')
print(X)

transformer = MaxAbsScaler().fit(X)

print('transformed data:')
print(transformer.transform(X))
```
